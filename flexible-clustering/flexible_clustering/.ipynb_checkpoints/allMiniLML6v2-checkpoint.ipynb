```python
# Use the 'message' field for clustering
messages = df['message'].values
```

```python
# Use the 'message' field for clustering
inputs = df['input'].values
```

```python
# Define Levenshtein distance
def levenshtein_distance(x, y):
    return Levenshtein.distance(x, y)
```

```python
def tlsh_distance(x, y):
    if len(x.strip()) <= 60 or len(y.strip()) <= 60:
        # print('here', x, y)
        return levenshtein_distance(x, y)
    # print(x,y)
    # Encode strings to bytes before passing them to TLSH
    hash_x = tlsh.hash(x.encode('utf-8'))  # Encoding to bytes
    hash_y = tlsh.hash(y.encode('utf-8'))  # Encoding to bytes
    
    # Calculate the distance between the two TLSH hashes
    # The distance method returns a value in the range [0, 100]
    return tlsh.diff(hash_x, hash_y)

# Initialize FISHDBC with TLSH-based distance function
fishdbc = FISHDBC(tlsh_distance)

# Apply FISHDBC to the entire dataset
fishdbc.update(inputs)  # Updating FISHDBC with the entire dataset at once
_, _, _, ctree, _, _ = fishdbc.cluster()

# Organize clusters
clusters = collections.defaultdict(set)
for parent, child, lambda_val, child_size in ctree[::-1]:
    if child_size == 1:
        clusters[parent].add(child)
    else:
        assert len(clusters[child]) == child_size
        clusters[parent].update(clusters[child])

clusters = sorted(clusters.items())

# Output the clustering results, showing the actual messages in each cluster
print("Clusters found:")
for parent, cluster in clusters:
    print(f"\nCluster {parent}:")
    for index in cluster:
        print(f"  - {inputs[index]}")  # Print the actual message in each cluster
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import collections

# Your original function and clustering logic remain unchanged
def tlsh_distance(x, y):
    if len(x.strip()) <= 60 or len(y.strip()) <= 60:
        return levenshtein_distance(x, y)
    
    hash_x = tlsh.hash(x.encode('utf-8'))  # Encoding to bytes
    hash_y = tlsh.hash(y.encode('utf-8'))  # Encoding to bytes
    
    return tlsh.diff(hash_x, hash_y)

# Initialize FISHDBC with TLSH-based distance function
fishdbc = FISHDBC(tlsh_distance)

# Apply FISHDBC to the entire dataset
fishdbc.update(messages)  # Updating FISHDBC with the entire dataset at once
_, _, _, ctree, _, _ = fishdbc.cluster()

# Organize clusters
clusters = collections.defaultdict(set)
for parent, child, lambda_val, child_size in ctree[::-1]:
    if child_size == 1:
        clusters[parent].add(child)
    else:
        assert len(clusters[child]) == child_size
        clusters[parent].update(clusters[child])

# Sort clusters in reverse order of size (so that the smallest clusters are at the end)
clusters = sorted(clusters.items())

# Select the bottom 10 clusters by size (smallest or most granular clusters)
bottom_clusters = clusters

# Create a graph to visualize the hierarchical structure
G = nx.DiGraph()

# Add nodes and edges to the graph
for parent, cluster in bottom_clusters:
    # Use the first message in the cluster as the representative
    representative_message = messages[next(iter(cluster))]  # Use the first element in the cluster as the representative
    
    # Check if parent is valid and not a root node (e.g., -1 or any invalid index)
    if parent != -1 and parent < len(messages):
        parent_message = messages[parent]
        G.add_edge(parent_message, representative_message)  # Create an edge from parent to representative message
        
    else:
        # If the parent is invalid (e.g., -1), treat the representative message as a root node
        G.add_node(representative_message)

# Draw the tree (hierarchical structure)
plt.figure(figsize=(12, 8))

# Use Graphviz to create a hierarchical layout for the tree structure
pos = nx.nx_agraph.graphviz_layout(G, prog="dot")  # This gives a more hierarchical tree layout

# Draw the graph using the calculated positions
nx.draw(G, pos, with_labels=True, node_size=5000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True)

# Display the plot
plt.title("Bottom 10 Clusters (Hierarchical Tree)")
plt.show()
```
