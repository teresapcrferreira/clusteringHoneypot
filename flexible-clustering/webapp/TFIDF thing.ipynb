```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option('display.max_colwidth', None)

# Load your table
df = pd.read_csv("../../discarded/dictCommands.csv")

# Combine relevant text fields
df['combined_description'] = (
    df['simplified_purpose'].fillna('') + ' ' +
    df['flag_description'].fillna('')
)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined_description'].values)

# Cosine similarity
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

# Create a labeled DataFrame for easy interpretation
similarity_df = pd.DataFrame(
    cosine_sim_matrix,
    index=df['label'],
    columns=df['label']
)
# # Optional: Save or explore
similarity_df.to_csv("finalissimoSim.csv")
```

```python
similarity_df = similarity_df.round(1)
# similarity_df
# Move wget rows to the top
wget_df = similarity_df[similarity_df['label'].str.startswith('wget')]
other_df = similarity_df[~similarity_df['label'].str.startswith('wget')]

# Concatenate them back together
df = pd.concat([wget_df, other_df], ignore_index=True)
df
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
import nltk

# Download stopwords if needed
nltk.download('stopwords')

# Load DataFrame
df = pd.read_csv("finalissimoCommands.csv")

# Fill NaNs with empty strings
df['flag_description'] = df['flag_description'].fillna('')
df['simplified_purpose'] = df['simplified_purpose'].fillna('')

# Define stop words and cleaner
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    words = text.split()  # Simple whitespace tokenizer
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return ' '.join(words)

# Clean both columns in place
df['flag_description'] = df['flag_description'].apply(clean_text)
df['simplified_purpose'] = df['simplified_purpose'].apply(clean_text)

# Preview result
df
```
